```python
import os
import pandas as pd
from datetime import date
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweets_list1 = []
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:PMOIndia since:2020-01-01 until:2023-01-01').get_items()):
    tweets_list1.append([tweet.date, tweet.content, tweet.id, tweet.user.username, tweet.likeCount,tweet.retweetCount,tweet.replyCount])
```

```python
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Text', 'Tweet Id', 'Username','likeCount','retweetCount','replyCount'])
tweets_df1
```

```python
tweets_df1.to_csv("data.csv", index=False,encoding = 'utf-8-sig')
```
